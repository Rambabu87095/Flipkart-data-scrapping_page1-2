```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY"
```

```python
res=requests.get(url)
```

```python
content=BeautifulSoup(res.content,'html.parser')
```

```python
print(content)
```

```python
data=content.find_all('div',{'class':'_2kHMtA'})
```

```python
title=content.find_all('div',{'class':'_4rR01T'})
rating=content.find_all('div',{'class':'_3LWZlK'})
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
```

```python
link=[]
start_link="https://www.flipkart.com"
for items in data:
    rest_link=items.find('a')['href']
    link.append(start_link+rest_link)
```

```python
Title=[]
Rating=[]
Price=[]
for i in title:
    Title.append(i.text)
for i in rating:
    Rating.append(i.text)
for i in price:
    Price.append(i.text)
```

```python
dataframe={'Title':Title,'Rating':Rating[:24],'Price':Price}
page1=pd.DataFrame(dataframe)
```

```python
url="https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page=2"
```

```python
res=requests.get(url)
```

```python
content=BeautifulSoup(res.content,'html.parser')
```

```python
data=content.find_all('div',{'class':'_2kHMtA'})
```

```python
title=content.find_all('div',{'class':'_4rR01T'})
rating=content.find_all('div',{'class':'_3LWZlK'})
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
```

```python
link=[]
start_link="https://www.flipkart.com"
for items in data:
    rest_link=items.find('a')['href']
    link.append(start_link+rest_link)
```

```python
Title=[]
Rating=[]
Price=[]
for i in title:
    Title.append(i.text)
for i in rating:
    Rating.append(i.text)
for i in price:
    Price.append(i.text)
```

```python
dataframe={'Title':Title,'Rating':Rating[:24],'Price':Price}
page2=pd.DataFrame(dataframe)
```
